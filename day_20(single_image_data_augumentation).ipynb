```python
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

```python
img=image.load_img(r"C:\Users\ASUS\Desktop\100_days_of_deeplearning\day_19\IMG_2132.JPG",target_size=(200,200))
```

```python
img
```

```python
datagen=ImageDataGenerator(
    rotation_range=40,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
)
```

```python
img=image.img_to_array(img)
```

```python
img
```

```python
img.shape
```

```python
input_batch=img.reshape(1,200,200,3)
```

```python
input_batch
```

```python
i=0
for output in datagen.flow(input_batch,batch_size=1,save_to_dir='augment'):
    i=i+1
    
    if i==10:
        break
```
