```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\ASUS\Desktop\100_days_of_deeplearning\day_16\concertriccir2.csv")
```

```python
df.head()
```

```python
plt.scatter(df['X'],df['Y'],c=df['class'])
```

```python
X=df.iloc[:,0:2].values
y=df.iloc[:,-1].values
```

```python
#without using batch_normalization
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
from keras.layers import BatchNormalization
```

```python
model=Sequential()
model.add(Dense(2,input_dim=2,activation='relu'))
model.add(Dense(2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history=model.fit(X,y,validation_split=0.2,epochs=100)
```

```python
#using a batch normalization
model1=Sequential()
model1.add(Dense(2,input_dim=2,activation='relu'))
model1.add(BatchNormalization())
model1.add(Dense(2,activation='relu'))
model1.add(BatchNormalization())
model1.add(Dense(1,activation='sigmoid'))
```

```python
model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history2=model1.fit(X,y,validation_split=0.2,epochs=100)
```

```python
#seeing the garph of both the accuracy
plt.plot(history.history['val_accuracy'],color='blue')
plt.plot(history2.history['val_accuracy'],color='red')
```
