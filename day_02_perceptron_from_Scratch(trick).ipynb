```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
X,y= make_classification(n_samples=200,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=41)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
X
```

```python
np.insert(X,0,1,axis=1)
```

```python
def Perceptron(X,y):
    X=np.insert(X,0,1,axis=1)
    weight=np.ones(X.shape[1])
    lr=0.1
    for i in range(1000):
        j=np.random.randint(0,100)
        y_hat=step(np.dot(X[j],weight))
        weight= weight+lr*(y[j]-y_hat)*X[j]
    return weight[0],weight[1:]
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept,coeff=Perceptron(X,y)
```

```python
intercept
```

```python
coeff
```

```python
m=-(coeff[0]/coeff[1])
```

```python
m
```

```python
b=-(intercept/coeff[1])
```

```python
b
```

```python
X_input= np.linspace(-3,3,100)
```

```python
y_input=m*X_input+b
```

```python
plt.figure(figsize=(10,5))
plt.plot(X_input,y_input,color='blue')
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter')
```
